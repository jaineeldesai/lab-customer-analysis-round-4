## Start of Lab 5

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
numeric.head()
```

```python
numeric.describe()
```

```python
sns.boxplot(numeric['Customer Lifetime Value'])
plt.show()
```

```python
sns.boxplot(numeric)
plt.xticks(rotation=45)
plt.show()
```

```python
#need to remove outlier from customer lifetime value
```

```python
iqr=np.percentile(numeric['Customer Lifetime Value'],75) - np.percentile(numeric['Customer Lifetime Value'],25)
iqr
```

```python
upper_limit = np.percentile(numeric['Customer Lifetime Value'],75) + 1.1*iqr
lower_limit = np.percentile(numeric['Customer Lifetime Value'],25) - 1.5*iqr
upper_limit
```

```python
num_data2 = numeric[(numeric['Customer Lifetime Value']>lower_limit) & (numeric['Customer Lifetime Value']<upper_limit)].copy()
sns.displot(num_data2['Customer Lifetime Value'])
plt.show()
```

```python
numeric.shape
```

```python
num_data2.shape
```

```python
num_data2.describe()
```

```python
sns.boxplot(num_data2['Customer Lifetime Value'])
plt.show()
```

```python
sns.boxplot(num_data2)
plt.show()
```

```python
y=num_data2['Total Claim Amount']
X=num_data2.drop(['Total Claim Amount'],axis=1)
```

```python
transformer = MinMaxScaler().fit(X)
x_normalized = transformer.transform(X)
print(x_normalized.shape)


x_norm=pd.DataFrame(x_normalized, columns=X.columns)
x_norm.head()
```

```python
sns.boxplot(x_norm)
plt.xticks(rotation=45)
plt.show()
```

```python
#cleared the outliers from customer lifetime value
```
